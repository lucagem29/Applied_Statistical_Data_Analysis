### Checking correlations 


```python
import pandas as pd
from IPython.display import display 

# Datei importieren
file_path = "Character types.xlsx"  # Der Pfad zur Datei

# Datei laden ohne automatische Datumsinterpretation
data = pd.read_excel(file_path, parse_dates=False)

# Überprüfen, welche Spalten betroffen sind
display(data.dtypes)  # Zeigt den Datentyp jeder Spalte an

# Beispiel: Falsche Datumswerte als Zahlen formatieren
# Konvertiere betroffene Spalten manuell, falls nötig
for col in data.columns:
    if pd.api.types.is_datetime64_any_dtype(data[col]):
        # Falsche Datumswerte in Zahlen umwandeln
        data[col] = data[col].astype(str).str.extract(r'(\d+\.\d+)')[0].astype(float)

# Fehlende Werte mit 0 ersetzen
data = data.fillna(0)

# Überprüfen, ob die Konvertierung korrekt ist
display(data.head(30))

```

```python
# NaN-Werte mit 0 ersetzen
data = data.fillna(0)
display(data.head(30))
```

```python
import pandas as pd

# Nur numerische Spalten auswählen
numeric_data = data.select_dtypes(include=['number'])

# Korrelationsmatrix berechnen
correlation_matrix = numeric_data.corr()

# Korrelationsmatrix anzeigen
print("Correlation Matrix:")
display(correlation_matrix)
```

```python
import pandas as pd
import numpy as np
from scipy.stats import pearsonr

# Nur numerische Spalten auswählen
numeric_data = data.select_dtypes(include=['number'])

# Initialisiere leere DataFrames für Korrelationen und p-Werte
correlation_matrix = pd.DataFrame(index=numeric_data.columns, columns=numeric_data.columns)
p_value_matrix = pd.DataFrame(index=numeric_data.columns, columns=numeric_data.columns)

# Berechnung der Korrelationswerte und p-Werte
for col1 in numeric_data.columns:
    for col2 in numeric_data.columns:
        if col1 == col2:
            # Selbstkorrelationen
            correlation_matrix.loc[col1, col2] = 1
            p_value_matrix.loc[col1, col2] = 0
        else:
            # Berechne Pearson-Korrelation und p-Wert
            corr, p_val = pearsonr(numeric_data[col1], numeric_data[col2])
            correlation_matrix.loc[col1, col2] = corr
            p_value_matrix.loc[col1, col2] = p_val

# Markiere signifikante Korrelationen (p < 0.05)
significance_matrix = (p_value_matrix < 0.05).astype(int)  # 1 = signifikant, 0 = nicht signifikant

# Ergebnisse anzeigen
print("Korrelationen:")
display(correlation_matrix)

print("Signifikanzmatrix (1 = signifikant, 0 = nicht signifikant):")
display(significance_matrix)
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Nur numerische Spalten auswählen
numeric_data = data.select_dtypes(include=['number'])

# Initialisiere leere DataFrames für Korrelationen und p-Werte
correlation_matrix = pd.DataFrame(index=numeric_data.columns, columns=numeric_data.columns, dtype=float)
p_value_matrix = pd.DataFrame(index=numeric_data.columns, columns=numeric_data.columns, dtype=float)

# Berechnung der Korrelationswerte und p-Werte
for col1 in numeric_data.columns:
    for col2 in numeric_data.columns:
        if col1 == col2:
            # Selbstkorrelationen
            correlation_matrix.loc[col1, col2] = 1
            p_value_matrix.loc[col1, col2] = 0
        else:
            # Berechne Pearson-Korrelation und p-Wert
            corr, p_val = pearsonr(numeric_data[col1], numeric_data[col2])
            correlation_matrix.loc[col1, col2] = corr
            p_value_matrix.loc[col1, col2] = p_val

# Heatmap erstellen
plt.figure(figsize=(12, 10))
sns.heatmap(
    correlation_matrix.astype(float),
    annot=True,                # Korrelationen als Text anzeigen
    fmt=".2f",                 # Format der Zahlen
    cmap="coolwarm",           # Farbskala
    mask=(p_value_matrix >= 0.05),  # Maskiere nicht-signifikante Werte
    cbar_kws={'label': 'Correlation'}
)

# Titel und Anzeige
plt.title("Heatmap der Korrelationen (nur signifikante Werte)", fontsize=16)
plt.show()
```
